```python
#TASK-1

class Calculator:
    def __init__(self):
        print('Letâ€™s Calculate!')
    def add(self,num_1,num_2):
        self.num_1=num_1
        self.num_2=num_2
        return 'Result: '+str(self.num_1+self.num_2)
    def subtract(self,num_1,num_2):
        self.num_1=num_1
        self.num_2=num_2
        return 'Result: '+str(self.num_1-self.num_2)
    def multiply(self,num_1,num_2):
        self.num_1=num_1
        self.num_2=num_2
        return 'Result: '+str(self.num_1*self.num_2)
    def divide(self,num_1,num_2):
        self.num_1=num_1
        self.num_2=num_2
        return 'Result: '+str(self.num_1/self.num_2)

call=Calculator()
number_1=int(input('Please enter first value:'))
operand=input('Please enter operand:')
number_2=int(input('Please enter second value:'))
if operand=='+':
    print(call.add(number_1,number_2))
elif operand=='-':
    print(call.subtract(number_1,number_2))
elif operand=='*':
    print(call.multiply(number_1,number_2))
else:
    print(call.divide(number_1,number_2))
```

```python
#TASK-21


class Hospital:
    def __init__(self,hospital_name):
        self.hospital_name=hospital_name
        self.doc_dict={}
        self.pat_dict={}
    def addDoctor(self,doctor):
        self.doc_dict[doctor.id]=[doctor.name,doctor.specialty]
    def getDoctorByID(self,id):
        print("Doctor's ID:",id)
        print('Name:',self.doc_dict[id][0])
        return 'Specialty: '+self.doc_dict[id][1]
    def addPatient(self,patient):
        self.pat_dict[patient.p_id]=[patient.p_name,patient.p_age,patient.p_number]
    def getPatientByID(self,p_id):
        print("Patient's ID:",p_id)
        print('Name:',self.pat_dict[p_id][0])
        print('Age:',self.pat_dict[p_id][1])
        return 'Phone no.: '+str(self.pat_dict[p_id][2])
    def allDoctors(self):
        print('All Doctors:')
        print('Number of Doctors:',len(self.doc_dict))
        print(self.doc_dict)
    def allPatients(self):
        print("All Patients:")
        print('Number of Patients:',len(self.pat_dict))
        print(self.pat_dict)
class Doctor:
    def __init__(self,id,occupation,name,specialty):
        self.id=id
        self.occupation=occupation
        self.name=name
        self.specialty=specialty
class Patient:
    def __init__(self,p_id,p_occupation,p_name,p_age,p_number):
        self.p_id=p_id
        self.p_occupation=p_occupation
        self.p_name=p_name
        self.p_age=p_age
        self.p_number=p_number
h = Hospital("Evercare")
d1 = Doctor("1d","Doctor", "Samar Kumar", "Neurologist")
h.addDoctor(d1)
print("=================================")
print(h.getDoctorByID("1d"))
print("=================================")
p1 = Patient("1p","Patient", "Kashem Ahmed", 35, 12345)
h.addPatient(p1)
print("=================================")
print(h.getPatientByID("1p"))
print("=================================")
p2 = Patient ("2p","Patient", "Tanina Haque", 26, 33456)
h.addPatient(p2)
print("=================================")
print(h.getPatientByID("2p"))
print("=================================")
h.allDoctors()
h.allPatients()
```

```python
#TASK-22

class Vaccine:
    def __init__(self,name,country,day):
        self.name=name
        self.country=country
        self.day=day
class Person:
    def __init__(self,person_name,age,occupation='General Citizen'):
        self.person_name=person_name
        self.age=age
        self.occupation=occupation
        self.vac=''
    def pushVaccine(self,vaccine_name,dose='1st dose'):
        self.first_dose=False
        self.second_dose=False
        if dose=='1st dose':
            if self.age>=25 or self.occupation=='Student':
                self.vac=vaccine_name
                print('1st dose done for',self.person_name)
                self.first_dose=True
            else:
                print('Sorry',self.person_name+',','Minimum age for taking vaccines is 25 years now.')
        else:
            if self.vac.name!=vaccine_name.name:
                print('Sorry',self.person_name+',','you canâ€™t take 2 different vaccines')
            else:
                print('2nd dose done for',self.person_name)
                self.second_dose=True
    def showDetail(self):
        print('Name:',self.person_name,'Age:',self.age,'Type:',self.occupation)
        print('Vaccine name:',self.vac.name)
        if self.second_dose==True:
            print('1st dose: Given')
            print('2nd dose: Given')
        else:
            print('1st dose: Given')
            print('2nd dose: Please come after',str(self.vac.day),'days')

astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
